
<br>
Series is a one-dimensional labeled array capable of holding data of any type (integer,<br>
string, float, python objects, etc.). The axis labels are collectively called index.<br>
pandas.Series<br>
A pandas Series can be created using the following constructor:<br>
pandas.DataFrame( data, index, dtype, copy)<br>
The parameters of the constructor are as follows:<br>
    data : data takes various forms like ndarray, list, constants<br>
    index : Index values must be unique and hashable, same length as data.<br>
    Default np.arrange(n) if no index is passed.<br>
    dtype : dtype is for data type. If None, data type will be inferred<br>
    copy : Copy data. Default False<br>
A series can be created using various inputs like:<br>
    Array<br>
    Dict<br>
    Scalar value or constant<br>


```python
import pandas as pd
import numpy as np
```

Create an Empty Series

```python
s =pd.Series()
print(s)
```

Create a Series from ndarray<br>
If no index is passed,then by default index will be range(n) where n is array length, i.e., [0,1,2,3â€¦.range(len(array))-1]

```python
data = np.array(['a','b','c','d','e'])
s = pd.Series(data)
print(s)
```

import the pandas library and aliasing as pd

```python
data = np.array(['a','b','c','d'])
s = pd.Series(data,index=[101,102,103,104])
print(s)
```

Create a Series from dict

```python
data = {'age':50 , 'name':'Dipika', 'occ':'home maker'}
s = pd.Series(data)
print(s)
```
